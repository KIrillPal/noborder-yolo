```python
import pandas as pd
import pickle
```

```python
all_results = []
with open(f'box_union_gridsearch.pkl', 'rb') as f:
    all_results.append(pickle.load(f))
with open(f'box_nms_gridsearch.pkl', 'rb') as f:
    all_results.append(pickle.load(f))
with open(f'box_softnms_gridsearch.pkl', 'rb') as f:
    all_results.append(pickle.load(f))
    
train_results = {}
for r in all_results:
    for k, v in r.items():
        k[0] = 'box'
        train_results[k] = v
```

```python
def sym(flag):
    return '+' if flag else ' '
```

```python
def get_table(train_results, test_results):
    table_list = []
    for border_filter in [False, True]:
        for threshold_filter in [False, True]:
            for conf_filter in [False, True]:
                for merger_type in ['Union', 'NMS', 'Soft-NMS']:
                    best_k = None
                    best_v = None
                    best_f1 = 0.0
                    for k, v in train_results.items():
                        if k[2] != border_filter:
                            continue
                        if (k[4] != 1) != threshold_filter:
                            continue
                        if (k[5] > k[3]) != conf_filter:
                            continue
                        if k[1] != merger_type.lower():
                            continue
                        if v[0] > best_f1:
                            best_k = k
                            best_v = v
                            best_f1 = v[0]
                    v = test_results[best_k]
                    table_list.append([
                        sym(border_filter), 
                        sym(threshold_filter), 
                        sym(conf_filter), 
                        merger_type, 
                        round(v[0], 2), 
                        round(v[1], 2), 
                        round(v[2], 2)
                    ])
    df = pd.DataFrame(table_list, columns=["B filter", "T filter", "C filter", "Merge type", "F1", "P", "R"])
    return df
```

```python

with open(f'box_sparse_gridsearch.pkl', 'rb') as f:
    test_results = pickle.load(f)
get_table(train_results, test_results)
```

```python
with open(f'box_dense_gridsearch.pkl', 'rb') as f:
    test_results = pickle.load(f)
get_table(train_results, test_results)
```
