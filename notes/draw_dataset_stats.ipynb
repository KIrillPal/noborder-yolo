```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import numpy as np

def draw_density_hist(data, bins=15):
    plt.clf()
    plt.hist(data, bins=bins, edgecolor="black", weights=np.ones(len(data)) / len(data))

    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    plt.xlim(0, 1)
    plt.title("Разреженный тест", fontsize=14)
    plt.xlabel("Плотность потока", fontsize=13)
    plt.ylabel("Доля изображений", fontsize=13)
    plt.show()
```

```python
import pandas as pd

sparse_data = pd.read_csv('../data/sparse_test/density/densities.csv').to_numpy()
dense_data = pd.read_csv('../data/dense_test/density/densities.csv').to_numpy()
```

```python
draw_density_hist(sparse_data)
```

```python
draw_density_hist(dense_data)
```
